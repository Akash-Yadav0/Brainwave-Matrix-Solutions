# Product Price Distribution

```python
sns.histplot(df["retail_price"], kde=True, bins=50, color='skyblue')
plt.title("Retail Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

# Discount Comparison (Retail vs Discounted Price)

```python
sample = df.sample(500)  # Limit for clarity
plt.figure(figsize=(8,5))
sns.scatterplot(x="retail_price", y="discounted_price", data=sample)
plt.title("Retail vs Discounted Price")
plt.xlabel("Retail Price")
plt.ylabel("Discounted Price")
plt.show()
```

# Top Brands by Product Count

```python
top_brands = df["brand"].value_counts().head(10)
plt.figure(figsize=(8,5))
top_brands.plot(kind="bar", color="coral")
plt.title("Top 10 Brands By Product Count")
plt.xlabel("Brand")
plt.ylabel("No. of Products")
plt.xticks(rotation=45)
plt.show()
```

# Category-wise Product Count

```python
df["main_category"] = df["product_category_tree"].apply(lambda x: eval(x)[0].split(">>")[0].strip())
top_cats = df["main_category"].value_counts().head(10)

top_cats.plot(kind="barh", color="lightgreen")
plt.title("Top 10 Product Categories")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

#  Rating Distribution (if numeric)

```python
df["numeric_rating"] = pd.to_numeric(
    df["overall_rating"].str.extract(r"(\d+\.?\d*)")[0],
    errors='coerce'
)

# Drop rows where numeric_rating is NaN
ratings = df["numeric_rating"].dropna()

# Plot
sns.boxplot(x=ratings)
plt.title("Boxplot of Overall Product Ratings")
plt.xlabel("Rating")
plt.show()
```

# Discount Percentage Distribution

```python
df["discount_percent"] = ((df["retail_price"] - df["discounted_price"]) / df["retail_price"]) * 100
sns.histplot(df["discount_percent"], bins=50, color="purple", kde=True)
plt.title("Discount Percentage Distribution")
plt.xlabel("Discount %")
plt.show()
```
